```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[[3.6,4.0,5.0,5.2,4.9],
   [3.9,4.6,5.6,5.9,5.8],
   [4.0,5.0,6.0,6.1,6.0],
   [4.1,5.4,5.8,6.5,6.1],]
y=[1,2,4,6,10]
x=np.array(x)
y=np.array(y)

plt.figure(0)
color=['r','g','y','b']
for i in range(len(x)):
    plt.plot(y,x[i],color=color[i],marker='o',label='rank '+str(i+1))
font={'family':'Times New Roman',
      'size':14}
plt.xlabel('cooperation intensity '+r'$\phi_0:\phi_1$',fontdict=font)
plt.ylabel('total reward / 10w',fontdict=font)
plt.legend()
plt.title('Affect of cooperation intensity in different scope')
```

```python
x= [3.1,5.1,6.5,6.8,6.8,6.6,5.7]
y= [0,1,2,3,4,5,6]
plt.figure(0)
plt.plot(y,x,color='r',marker='o')
font={'family':'Times New Roman',
      'size':14}
plt.xlabel('cooperation scope',fontdict=font)
plt.ylabel('total reward / 10w',fontdict=font)
plt.title('Affect of different cooperation scope')
```
